# Support Vector Regression (SVR) and Support Vector Machine (SVM):

```python
1].Purpose:
    
SVM: Primarily used for classification tasks, where the goal is to find a hyperplane 
     that separates different classes in the feature space.
        
SVR: Used for regression tasks, aiming to find a function that approximates the relationship between
     input variables and a continuous target variable.

2].Output:

SVM: Produces discrete class labels as output.
    
SVR: Produces continuous values as output.
    
3].Objective:
    
SVM: Maximizes the margin between the closest data points of different classes (support vectors) and the hyperplane.
    
SVR: Minimizes the error within a specified margin (epsilon) while also trying to keep the model as flat as possible.
    
4].Loss Function:
    
SVM: Uses hinge loss for classification, which penalizes misclassified points.

SVR: Uses epsilon-insensitive loss, which ignores errors within a certain margin and penalizes errors outside this margin.

5].Hyperparameters:
    
SVM: Key hyperparameters include the regularization parameter C and the kernel type.
    
SVR: Similar to SVM, but also includes the epsilon parameter () that defines the margin of tolerance for errors
```
